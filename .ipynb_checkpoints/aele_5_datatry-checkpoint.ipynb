```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def visualize_vowel_points(df, normalized=False, mode="2D"):
    """
    Visualize vowel Gaussian data.

    Parameters
    ----------
    df : pandas.DataFrame
        Must contain columns: ['vowel', 'F3', 'F2', 'F1']
        or their normalized versions ['F3_norm', 'F2_norm', 'F1_norm'].
    normalized : bool, default=False
        Whether to plot normalized columns (F*_norm) instead of raw.
    mode : {'2D', '3D'}, default='2D'
        Choose 2D F1–F2 scatter or full 3D F1–F2–F3 visualization.
    """

    # choose which columns to plot
    if normalized:
        F3, F2, F1 = "F3_norm", "F2_norm", "F1_norm"
        title_suffix = " (log-mean normalized)"
    else:
        F3, F2, F1 = "F3", "F2", "F1"
        title_suffix = ""

    vowels = sorted(df["vowel"].unique())
    colors = plt.cm.tab10.colors  # nice categorical palette
    color_map = {v: colors[i % len(colors)] for i, v in enumerate(vowels)}

    if mode == "3D":
        fig = plt.figure(figsize=(8, 6))
        ax = fig.add_subplot(111, projection="3d")
        for v in vowels:
            subset = df[df["vowel"] == v]
            ax.scatter(subset[F2], subset[F1], subset[F3],
                       s=10, alpha=0.6, label=v, color=color_map[v])
        ax.set_xlabel("F2")
        ax.set_ylabel("F1")
        ax.set_zlabel("F3")
        ax.set_title(f"3D vowel space{title_suffix}")
        ax.legend()
        plt.show()

    else:  # 2D view (F1–F2)
        plt.figure(figsize=(7, 6))
        for v in vowels:
            subset = df[df["vowel"] == v]
            plt.scatter(subset[F2], subset[F1],
                        s=10, alpha=0.6, label=v, color=color_map[v])
        plt.xlabel("F2")
        plt.ylabel("F1")
        plt.gca().invert_yaxis()  # conventional vowel chart orientation
        plt.title(f"Vowel space (F1–F2){title_suffix}")
        plt.legend()
        plt.tight_layout()
        plt.show()
```

```python
# -----------------------------
# Configuration
# -----------------------------
np.random.seed(42)

output_dir = "./vowel_gaussian_data"
os.makedirs(output_dir, exist_ok=True)

vowel_list = ['i', 'L', 'e', 'F', 'u', 'W', 'o', 'D']
vowel_formants = [
    [3372, 2761, 437],
    [3053, 2365, 483],
    [3047, 2530, 536],
    [2979, 2058, 731],
    [2735, 1105, 459],
    [2827, 1225, 519],
    [2828, 1035, 555],
    [2824, 1136, 781]
]

# Standard deviation as a proportion of mean (e.g., 5%)
std_ratio = 0.05
n_samples_per_vowel = 1000

# -----------------------------
# 1. Generate Gaussian data
# -----------------------------
records = []
for vowel, means in zip(vowel_list, vowel_formants):
    means = np.array(means)
    stds = means * std_ratio

    # Generate n_samples × 3 matrix (F3, F2, F1)
    samples = np.random.normal(loc=means, scale=stds, size=(n_samples_per_vowel, 3))
    for s in samples:
        records.append({
            "vowel": vowel,
            "F3": s[0],
            "F2": s[1],
            "F1": s[2]
        })

df_raw = pd.DataFrame(records)
raw_path = os.path.join(output_dir, "raw_vowel_formants.csv")
df_raw.to_csv(raw_path, index=False)
print(f"Saved raw data to {raw_path}")

# -----------------------------
# 2. Per-dimension log-mean normalization
# -----------------------------
def logmean_normalize(X):
    """
    Per-dimension log-mean normalization:
    X: numpy array of shape (n_samples, n_dims)
    returns normalized data, per-dim logmean, logstd
    """
    eps = 1e-8
    X_logged = np.log(X + eps)
    log_means = np.mean(X_logged, axis=0, keepdims=True)
    log_stds = np.std(X_logged, axis=0, keepdims=True)
    X_norm = (X_logged - log_means) / (log_stds + eps)
    return X_norm, log_means.flatten(), log_stds.flatten()

formant_data = df_raw[["F3", "F2", "F1"]].values
formant_norm, log_means, log_stds = logmean_normalize(formant_data)

df_norm = df_raw.copy()
df_norm[["F3_norm", "F2_norm", "F1_norm"]] = formant_norm

norm_path = os.path.join(output_dir, "normalized_vowel_formants.csv")
df_norm.to_csv(norm_path, index=False)
print(f"Saved normalized data to {norm_path}")

# -----------------------------
# 3. Save normalization summary
# -----------------------------
summary = pd.DataFrame({
    "dimension": ["F3", "F2", "F1"],
    "logmean": log_means,
    "logstd": log_stds
})
# summary_path = os.path.join(output_dir, "logmean_summary.csv")
# summary.to_csv(summary_path, index=False)
# print(f"Saved log-mean normalization summary to {summary_path}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

# Parameters
tau = 1/0.001          # time constant
N = 1           # number of units
gamma = 0.5        # learning rate scaling
w0 = 1e-7           # initial weight
shift = 10000
base = 0
t = np.linspace(base, base+shift, 500)  # time vector (avoid t=0 to prevent singularity)

# Range of lambda_i values
n = 3
# lambdas = N*(np.random.rand(n))/((0.2-0))
lambdas_raw = [2.5, 2.0, 0.2]
lambdas = []
for l in lambdas_raw: 
    lambdas.append(l*N)
n = len(lambdas)

def w_i(t, lambda_i, tau, N, gamma, w0):
    xi = 1 - (N * gamma) / lambda_i
    num = xi * np.exp((2 * xi * lambda_i / tau) * t)
    den = np.exp((2 * xi * lambda_i / tau) * t) - 1 + xi / w0
    return num / den

# Plot
plt.figure(figsize=(7, 5))
ws = []
# w_x = w_i(t, 0.15, tau, N, gamma, w0)
w_total = 0
for lam in lambdas:
    w = w_i(t, lam, tau, N, gamma, w0)
    # ws[lam] = w
    ws.append(w)
    w_total += w
    
for i, w in enumerate(ws):    
    # plt.plot(t, w, label=f"$λ_i$=")
    plt.plot(t, w - (w_total/n), label=f"$λ_i$={lambdas[i]}")

plt.xlabel("t")
plt.ylabel("$w_i$")
plt.title("Weight dynamics for different λ")
plt.legend()
plt.grid(True)
plt.tight_layout()
# plt.show()
plt.savefig('./writing_plot/theory_curve_relative.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
tau = 1/0.0005          # time constant
N = 1
gamma = 0.1
w0 = 1e-7
shift = 1_000_000
base = 0
t = np.linspace(base, base + shift, 500)

# Lambdas
lambdas = [0.12*N, 0.15*N, 0.18*N]

def w_i(t, lambda_i, tau, N, gamma, w0):
    xi = 1 - (N * gamma) / lambda_i
    a = (2 * xi * lambda_i / tau)  # effective rate
    num = xi * np.exp(a * t)
    den = np.exp(a * t) - 1 + xi / w0
    return num / den  # w(0)=w0; w(∞)=xi

# Compute all curves
W = []
for lam in lambdas:
    W.append(w_i(t, lam, tau, N, gamma, w0))
W = np.stack(W, axis=0)  # shape (C, T)

# RDI: divide each curve by per-epoch mean across lambdas
eps = 1e-12
mean_per_epoch = W.mean(axis=0)
denom = np.maximum(mean_per_epoch, eps)
RDI = W / denom  # broadcasting

# Plot
plt.figure(figsize=(7, 5))
for lam, rdi_curve in zip(lambdas, RDI):
    plt.plot(t, rdi_curve, label=fr"$\lambda$={lam:g}")
# plt.plot(t, np.ones_like(t), 'k--', lw=1.2, label="epoch mean (RDI=1)")

plt.xlabel("t")
plt.ylabel("RDI = wλ(t) / meanλ w(t)")
plt.title("Theoretical curves with per-epoch RDI normalization")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
