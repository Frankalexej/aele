# Figure 5: MNIST and CIFAR-10 theoretical DAE/WDAE learning dynamics compared to actual learning dynamics.

This notebook provides the code to produce Figure 5 in the paper: "Learning dynamics of linear denoising autoencoders". (ICML 2018)

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
from torchvision import transforms
from torchvision.datasets import MNIST
from torchvision.datasets import CIFAR10
from collections import OrderedDict

# custom imports
from src.linear_ae_net.linear_ae_net import LinearAutoEncoder
from src.linear_ae_net.dynamics import *
```

## --- MNIST ---

### Load MNIST data

```python
# cast to tensor
trans = transforms.Compose([transforms.ToTensor()])

# if not exist, download mnist dataset
train_set = MNIST(root="../data", train=True, transform=trans, download=True)
x_train = train_set.train_data.numpy()
x_train = x_train.astype('float32') / 255.
x_train_mnist = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
print(x_train_mnist.shape)
```

### Train models

```python
# set parameters
num_samples = 50000
epochs = 5000
lr = 0.01
reg_param = 0.5
var_param = 0.5
reg = [0.0, reg_param, 0.0]
var = [0.0, 0.0, var_param]
num_trials = 3
hidden_dim = 256

mnist_models = []

# convert to pytorch tensors
x_train_mnist = torch.from_numpy(x_train_mnist)

# set seed
np.random.seed(123)
torch.manual_seed(321)

# train autoencoder network
for t in range(num_trials):
    laeModel = LinearAutoEncoder()
    laeModel.train(x_train_mnist[:num_samples,], None, input_dim=784, n_epoch=epochs, 
                   hidden_dim=hidden_dim, learning_rate=lr, reg_param=reg[t], 
                   noise='Gaussian', noise_scale=var[t], verbose=True)
    mnist_models.append(laeModel)
```

### Compute theoretical dynamics

```python
def theoretical_learning_dynamics_here(X, y, n_epoch, lr, var, reg, u0 = 2.5e-7):
    dyns = np.zeros((1,X.shape[1]))
    corr_mat = compute_correlation_matrix(X, y)
    U, S, V = np.linalg.svd(corr_mat)
    dyns = dae_learning_dynamics(S, var, reg, X.shape[0], n_epoch, lr, u0)
    return dyns, corr_mat, U, S, V
```

```python
# compute MNIST dynamics
x_train_mnist_np = x_train_mnist.cpu().numpy()
w0 = mnist_models[0].init
res = theoretical_learning_dynamics_here(
                                x_train_mnist_np[:num_samples, :], 
                                x_train_mnist_np[:num_samples, :], 
                                n_epoch=epochs, lr=lr, 
                                var=0, reg=reg_param, u0=w0)
```

```python
theoretical_learning_dynamics_here, corr_mat, U, S, V = res
```

```python
plt.figure(figsize=(10, 6))
plt.plot(S)
plt.title('Change of Singular Values')
plt.xlabel('Index')
plt.ylabel('Singular Value')
plt.grid()
plt.show()
```

```python
import seaborn as sns

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_mat, cmap='coolwarm', annot=False, cbar=True)
plt.title('Heatmap of Correlation Matrix')
plt.xlabel('Features')
plt.ylabel('Features')
plt.show()
# corr_mat
```

```python

theoretical_dynamics_mnist = theoretical_learning_dynamics(
                                x_train_mnist_np[:num_samples, :], 
                                x_train_mnist_np[:num_samples, :], 
                                n_epoch=epochs, lr=lr, 
                                var=0, reg=0, u0=w0)
theoretical_dynamics_reg_mnist = theoretical_learning_dynamics(
                                x_train_mnist_np[:num_samples, :], 
                                x_train_mnist_np[:num_samples, :], 
                                n_epoch=epochs, lr=lr, 
                                var=0, reg=reg_param, u0=w0)
theoretical_dynamics_noise_mnist = theoretical_learning_dynamics(
                                x_train_mnist_np[:num_samples, :], 
                                x_train_mnist_np[:num_samples, :], 
                                n_epoch=epochs, lr=lr, 
                                var=var_param, reg=0, u0=w0)
```

## --- CIFAR-10 ---

### Load CIFAR-10 data

```python
trans = transforms.Compose([transforms.ToTensor()])
train_set = CIFAR10(root="../data", train=True, transform=trans, download=True)
x_train = train_set.train_data
x_train = x_train.astype('float32')
x_train /= 255
x_train_cifar10 = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
print('x_train shape:', x_train_cifar10.shape)
```

### Train models

```python
# set parameters
num_samples = 30000
epochs = 5000
lr = 0.001
reg_param = 0.5
var_param = 0.5
reg = [0.0, reg_param, 0.0]
var = [0.0, 0.0, var_param]
num_trials = 3
hidden_dim = 512

cifar10_models = []

# convert to pytorch tensors
x_train_cifar10 = torch.from_numpy(x_train_cifar10)

# set seed
np.random.seed(123)
torch.manual_seed(321)

# train autoencoder networks
for t in range(num_trials):
    laeModel = LinearAutoEncoder()
    laeModel.train(x_train_cifar10[:num_samples,], None, input_dim=32*32*3, 
                   n_epoch=epochs, hidden_dim=hidden_dim, 
                   learning_rate=lr, reg_param=reg[t], 
                   noise='Gaussian', noise_scale=var[t], verbose=True)
    cifar10_models.append(laeModel)
```

### Compute theoretical dynamics

```python
# compute CIFAR-10 dynamics
x_train_cifar10_np = x_train_cifar10.cpu().numpy()
w0 = cifar10_models[0].init
theoretical_dynamics_cifar10 = theoretical_learning_dynamics(
                                x_train_cifar10_np[:num_samples, :], 
                                x_train_cifar10_np[:num_samples, :], 
                                n_epoch=epochs, lr=lr, var=0, 
                                reg=0, u0 = w0)
theoretical_dynamics_reg_cifar10 = theoretical_learning_dynamics(
                                x_train_cifar10_np[:num_samples, :], 
                                x_train_cifar10_np[:num_samples, :], 
                                n_epoch=epochs, lr=lr, var=0, 
                                reg=reg_param, u0 = w0)
theoretical_dynamics_noise_cifar10 = theoretical_learning_dynamics(
                                x_train_cifar10_np[:num_samples, :], 
                                x_train_cifar10_np[:num_samples, :], 
                                n_epoch=epochs, lr=lr, var=var_param, 
                                reg=0, u0 = w0) 
```

```python
# create dynamics plot
slices = (0, 3, 7, 15, 31)
fig, [(ax1, ax2), (ax3, ax4)] = plt.subplots(2, 2, figsize=(12, 8), sharey='row', sharex='col')
axes = [ax1, ax2]

# plot theoretical dynamics
ax1.plot(theoretical_dynamics_mnist[:, slices], c='blue', 
         label='Theory ($\gamma = 0$)')
ax1.plot(theoretical_dynamics_reg_mnist[:, slices], c='orange', 
         label='Theory ($\gamma = $' + str(reg_param) + ')')
ax2.plot(theoretical_dynamics_mnist[:, slices], c='blue', 
         label='Theory ($\sigma^2 = 0$)')
ax2.plot(theoretical_dynamics_noise_mnist[:, slices], c='darkgreen', 
         label='Theory ($\sigma^2 = $' + str(var_param) + ')')

# get actual dynamics
actual_dynamics_mnist = mnist_models[0].strenghts.cpu().numpy()
actual_dynamics_reg_mnist = mnist_models[1].strenghts.cpu().numpy()
actual_dynamics_noise_mnist = mnist_models[2].strenghts.cpu().numpy()

# plot simulated dynamics
x_p = np.arange(0, epochs+1, 100)
for s in slices:
    ax1.scatter(x_p, actual_dynamics_mnist[:, s], c='blue', 
                marker='x', label='Actual ($\gamma = 0$)')
    ax1.scatter(x_p, actual_dynamics_reg_mnist[:, s], c='orange', 
                marker='x', label='Actual ($\gamma = $' + str(reg_param) + ')')
    ax2.scatter(x_p, actual_dynamics_mnist[:, s], c='blue', 
                marker='x', label='Actual ($\sigma^2 = 0$)')
    ax2.scatter(x_p, actual_dynamics_noise_mnist[:, s], c='darkgreen',
                marker='x', label='Actual ($\sigma^2 = $' + str(var_param) + ')')
    
    
# set plot titles and axis labels
ax1.set_ylabel('$w_2 \cdot w_1$', fontsize=15)
ax1.set_title('Weight decay')
ax2.yaxis.set_label_position('right')
ax2.set_ylabel('MNIST')

# remove duplicates labels
locations = ['lower right', 'lower right', 'upper right', 'upper right']
for ax, loc in zip(axes, locations):
    handles, labels = ax.get_legend_handles_labels()
    by_label = OrderedDict(zip(labels, handles))
    ax.legend(by_label.values(), by_label.keys(), loc=loc)

# plot theoretical dynamics
ax3.plot(theoretical_dynamics_cifar10[:, slices], c='blue')
ax3.plot(theoretical_dynamics_reg_cifar10[:, slices], c='orange')
ax4.plot(theoretical_dynamics_cifar10[:, slices], c='blue')
ax4.plot(theoretical_dynamics_noise_cifar10[:, slices], c='darkgreen')

# get actual dynamics
actual_dynamics_cifar10 = cifar10_models[0].strenghts.cpu().numpy()
actual_dynamics_reg_cifar10 = cifar10_models[1].strenghts.cpu().numpy()
actual_dynamics_noise_cifar10 = cifar10_models[2].strenghts.cpu().numpy()

# plot simulated dynamics
x_p = np.arange(0, epochs+1, 100)
for s in slices:
    ax3.scatter(x_p, actual_dynamics_cifar10[:, s], c='blue', marker='x')
    ax3.scatter(x_p, actual_dynamics_reg_cifar10[:, s], c='orange', marker='x')
    ax4.scatter(x_p, actual_dynamics_cifar10[:, s], c='blue', marker='x')
    ax4.scatter(x_p, actual_dynamics_noise_cifar10[:, s], c='darkgreen',marker='x')

# set plot titles and axis labels
ax2.set_title('Noise')
ax3.set_ylabel('$w_2 \cdot w_1$', fontsize=15)
ax3.set_xlabel('t (epoch)', fontsize=10)
ax4.set_xlabel('t (epoch)', fontsize=10)
ax4.yaxis.set_label_position('right')
ax4.set_ylabel('CIFAR-10')
plt.show()
```

```python
import numpy as np

def generate_gmm_data(num_data, d, n_components=3, random_state=None):
    """
    Generate a synthetic dataset drawn from a d-dimensional Gaussian Mixture Model (GMM).

    Parameters
    ----------
    num_data : int
        Number of data points to generate.
    d : int
        Dimension of each data point.
    n_components : int, optional (default=3)
        Number of mixture components.
    random_state : int or None
        Seed for the random number generator for reproducibility.

    Returns
    -------
    X : np.ndarray, shape (num_data, d)
        Generated data points.
    labels : np.ndarray, shape (num_data,)
        Component labels (0 to n_components-1) for each sample.
    """
    rng = np.random.default_rng(random_state)
    
    # 1. Mixture weights (sum to 1)
    weights = rng.random(n_components)
    weights /= weights.sum()
    
    # 2. Random means for each component
    means = rng.uniform(-10, 10, size=(n_components, d))
    
    # 3. Covariance matrices (we'll use axis-aligned Gaussians)
    #    For simplicity, variances are drawn between [0.5, 1.5]
    covariances = np.array([np.diag(rng.uniform(0.5, 1.5, size=d)) 
                            for _ in range(n_components)])
    
    # 4. Sample component assignments
    labels = rng.choice(n_components, size=num_data, p=weights)
    
    # 5. Allocate array for data
    X = np.empty((num_data, d))
    
    # 6. Draw samples
    for k in range(n_components):
        mask = (labels == k)
        count_k = mask.sum()
        if count_k > 0:
            X[mask] = rng.multivariate_normal(
                mean=means[k],
                cov=covariances[k],
                size=count_k
            )
    
    return X, labels
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_gmm_data(num_data, d, n_components=3, random_state=None):
    rng = np.random.default_rng(random_state)
    weights = rng.random(n_components)
    weights /= weights.sum()
    means = rng.uniform(-10, 10, size=(n_components, d))
    covariances = np.array([np.diag(rng.uniform(0.5, 1.5, size=d)) for _ in range(n_components)])
    labels = rng.choice(n_components, size=num_data, p=weights)
    X = np.empty((num_data, d))
    for k in range(n_components):
        mask = (labels == k)
        X[mask] = rng.multivariate_normal(mean=means[k], cov=covariances[k], size=mask.sum())
    return X, labels

# Generate a 2D GMM dataset
X, labels = generate_gmm_data(num_data=500, d=2, n_components=4)

# Scatter plot colored by component label
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('2D GMM Samples')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# If not already defined, regenerate the same 2D GMM data:
def generate_gmm_data(num_data, d, n_components=4, random_state=42):
    rng = np.random.default_rng(random_state)
    weights = rng.random(n_components)
    weights /= weights.sum()
    means = rng.uniform(-10, 10, size=(n_components, d))
    covariances = np.array([np.diag(rng.uniform(0.1, 0.5, size=d)) for _ in range(n_components)])
    labels = rng.choice(n_components, size=num_data, p=weights)
    X = np.empty((num_data, d))
    for k in range(n_components):
        mask = (labels == k)
        X[mask] = rng.multivariate_normal(means[k], covariances[k], size=mask.sum())
    return X, labels

X, labels = generate_gmm_data(num_data=500, d=2, n_components=3, random_state=calendar.timegm(time.gmtime()))

# 1) Covariance matrix heatmap
C = np.cov(X, rowvar=False)
plt.figure()
plt.imshow(C, aspect='equal')
plt.colorbar()
plt.title("Covariance Matrix")
plt.xlabel("Dimension")
plt.ylabel("Dimension")
plt.show()

# 2) Singular values
U, S, Vt = np.linalg.svd(C)
# plt.figure()
# plt.bar(np.arange(1, len(S)+1), S)
# plt.xlabel("Component")
# plt.ylabel("Singular Value")
# plt.title("Singular Values of Covariance Matrix")
# plt.show()

# 3) Scatter + singular vectors
mu = X.mean(axis=0)
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=labels)
for i in range(2):
    # scale arrow length by sqrt of singular value for visibility
    vec = U[:, i] * np.sqrt(S[i])
    plt.arrow(mu[0], mu[1], vec[0], vec[1], width=0.05, length_includes_head=True)
    plt.arrow(mu[0], mu[1], -vec[0], -vec[1], width=0.05, length_includes_head=True)
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.title("2D GMM Samples with Principal Axes")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Regenerate the same 2D GMM data
def generate_gmm_data(num_data, d, n_components=4, random_state=42, means=None):
    if means is not None and len(means) != n_components: 
        n_components = len(means)
    rng = np.random.default_rng(random_state)
    # weights = rng.random(n_components)
    weights = np.ones(n_components)
    weights /= weights.sum()
    if means is None: 
        means = rng.uniform(-20, 20, size=(n_components, d))
    else: 
        means = np.array(means)
    covariances = np.array([np.diag(rng.uniform(0.1, 0.2, size=d)) for _ in range(n_components)])
    labels = rng.choice(n_components, size=num_data, p=weights)
    X = np.empty((num_data, d))
    for k in range(n_components):
        mask = (labels == k)
        X[mask] = rng.multivariate_normal(means[k], covariances[k], size=mask.sum())
    return X, labels
```

```python
for num_pairs in range(1, 10): 
    # X, labels = generate_gmm_data(num_data=500, d=2)
    means = []
    for i in range(num_pairs): 
        means.append([-20, i*10])
        means.append([20, i*10])
    X, labels = generate_gmm_data(num_data=50000, d=2, n_components=4, random_state=calendar.timegm(time.gmtime()), means=means)
    # X, labels = generate_gmm_data(num_data=50000, d=2, n_components=3, random_state=42)

    # Compute covariance and SVD
    mu = X.mean(axis=0)
    C = np.cov(X, rowvar=False)
    U, S, Vt = np.linalg.svd(C)

    # Plot scatter
    plt.figure(figsize=(7, 7))
    plt.scatter(X[:, 0], X[:, 1], c=labels, alpha=0.6)

    # Draw and label principal axes
    for i in range(2):
        vec = U[:, i] * np.sqrt(S[i])  # vector direction and scale
        # Plot arrow
        plt.arrow(mu[0], mu[1], vec[0], vec[1],
                width=0.05, length_includes_head=True, label=f'σ{i+1}')
        plt.arrow(mu[0], mu[1], -vec[0], -vec[1],
                width=0.05, length_includes_head=True)
        # Annotate with singular value at arrow tip
        plt.text(mu[0] + vec[0]*1.1, mu[1] + vec[1]*1.1,
                f'{S[i]:.2f}', fontsize=12, fontweight='bold', ha='center', va='center')

    plt.xlabel("Dimension 1")
    plt.ylabel("Dimension 2")
    plt.title("2D GMM Samples with Principal Axes and Singular Values")
    plt.grid(True)
    # plt.show()
    plt.savefig(f'./plot_save/gmmsvd_{num_pairs}.png', dpi=300, bbox_inches='tight')
```
